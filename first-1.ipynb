Why Use with open(...) Instead of open(...)?
Using with open(...) is better than manually opening and closing a file because:

âœ… Automatic File Closing

The file is automatically closed after the block ends, even if an error occurs.

No need to write f.close() manually.

âœ… Prevents Resource Leaks

If you forget f.close(), your file remains open in memory, causing resource leaks.

âœ… Better Exception Handling

If an error occurs inside the with block, Python still closes the file properly.

How seek() Works
When a file is opened in read mode ('r'), the cursor starts at position 0.

The read(n) function moves the cursor forward by n characters.

The seek(position) method moves the cursor to the specified position, resetting where the next read starts



What is seek()?
ðŸ”¹ The seek(offset, whence) method is used to move the file pointer to a specific position in a file.

offset â†’ The number of bytes (characters) to move.

whence (optional) â†’ Defines the reference point for offset:

0 (default) â†’ Start from the beginning of the file.

1 â†’ Move relative to the current position.

2 â†’ Move relative to the end of the file.

How seek() Works Here?
f.read(10)

Reads the first 10 characters from the file.

Moves the file pointer to position 10.

f.seek(15)

Moves the pointer to position 15, skipping 5 characters.

f.read(13)

Reads 13 characters starting from position 15.

Why is seek() Needed?
It allows random access to a file, meaning we can jump to any position instead of reading from the beginning.

Useful when:

Skipping unnecessary data.

Reading a specific portion of a file.

Overwriting a part of the file.

```python
k = 0  # Initialize total

with open("sample1.txt", "w+") as f:
    for i in range(1, 11, 2):
        product = i * (i + 1)
        k += product
        line = f"{i}\t{i+1}\t{product}\n"
        f.write(line)

    # Write the total after the loop
    f.write(f"Total\t\t{k}\n")

# Read and display the contents
with open("sample1.txt", "r") as f:
    print(f.read())
```

Q-4: Create line wise reverse of a file
Write a function which takes two arguments: the names of the input file (to be read from) and the output file (which will be created).

For example, if a file looks like

abc def
ghi jkl
then the output file will be

fed cba
lkj ihg
Notice: The newline remains at the end of the string, while the rest of the characters are all reversed.

```python
def reverselines():
    file_name_taken = input("Enter the file name: ")
    generated_file_name = input("Enter the new file name: ")

    # Store the generated file name in a variable outside the function
    globals()['generated_file_name'] = generated_file_name

    with open(file_name_taken, "r") as f:
        lines_taken = f.readlines()

    with open(generated_file_name, "w") as f:
        for i in range(len(lines_taken) - 1, -1, -1):  # manual reverse
            f.write(lines_taken[i])

reverselines()

# Read and print contents of the new file
with open(generated_file_name, "r") as f:
    print(f.read())
```

```python
def reverselines():
    file_name_taken = input("Enter the file name: ")
    generated_file_name = input("Enter the new file name: ")

    with open(file_name_taken, "r") as f:
        lines_taken = f.readlines()

    with open(generated_file_name, "w") as f:
        for i in reversed(lines_taken):
            f.write(i)

    return generated_file_name  # Return the new filename

# Call the function and capture the output filename
output_file = reverselines()

# Display the reversed file contents
with open(output_file, "r") as f:
    print(f.read())
```
